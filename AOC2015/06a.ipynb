```python
class axis:
    def __init__(s,a,b):
        s.a=min(a,b)
        s.b=max(a,b)
    
    def length(s):
        return s.b-s.a
    
    def seperate(s, o):
        return s.a > o.b or s.b < o.a
    
    def touch(s,o):
        return not s.seperate(o)
    
    def dissolve(s, o):
        a = [] #s
        b = [] #union
        c = [] #o
        
        if s.a==o.a and s.b==o.b:
            #identical
            ######SSS######
            ######OOO######
            ######bbb######
            b.append(axis(s.a, s.b))
        elif s.seperate(o):
            #outside
            ###SSS#########
            ######OOO######
            ###aaaccc######
            a.append(axis(s.a, s.b))
            c.append(axis(o.a, o.b))
        elif s.a > o.a and s.b < o.b:
            #s inside o
            ######SSS######
            #####OOOOO#####
            #####cbbbc#####
            c.append(axis(o.a, s.a-1))
            b.append(axis(s.a, s.b))
            c.append(axis(s.b+1, o.b))
        elif o.a > s.a and o.b < s.b:
            #o inside s
            #####SSSSS#####
            ######OOO######
            #####abbba#####
            a.append(axis(s.a, o.a-1))
            b.append(axis(o.a, o.b))
            a.append(axis(o.b+1, s.b))
        elif s.a == o.a and s.b > o.b:
            #s extend o right
            #####SSSSS#####
            #####OOO#######
            #####bbbaa#####
            b.append(axis(s.a, o.b))
            a.append(axis(o.b+1, s.b))
        elif s.a == o.a and s.b < o.b:
            #o extend s right
            #####SSS#######
            #####OOOOO#####
            #####bbbcc#####
            b.append(axis(s.a, s.b))
            c.append(axis(s.b+1, o.b))
        elif s.a < o.a and s.b == o.b:
            #s extend o left
            ###SSSSS#######
            #####OOO#######
            ###aabbb#######
            a.append(axis(s.a, o.a-1))
            b.append(axis(o.a, o.b))
        elif s.a > o.a and s.b == o.b:
            #o extend s left
            #####SSS#######
            ###OOOOO#######
            ###ccbbb#######
            c.append(axis(o.a, s.a-1))
            b.append(axis(s.a, o.b))
        elif s.a > o.a and s.b > o.b:
            c.append(axis(o.a, s.a-1))
            b.append(axis(s.a, o.b))
            a.append(axis(o.b+1, s.b))
        elif s.a < o.a and s.b < o.b:
            a.append(axis(s.a, o.a-1))
            b.append(axis(o.a, s.b))
            c.append(axis(s.b+1, o.b))
        else:
            print("this should not happen")
        
        return (a,b,c)
    
    def print(s, c):
        l = ""
        for i in range(20):
            if i >= s.a and i <= s.b:
                l += c
            else:
                l += " "
        print(l)
                       
A = axis(3,7)
B = axis(5,8)
A.print("A")
B.print("B")
a,b,c = A.dissolve(B)
for x in a:
    x.print("a")
for x in b:
    x.print("b")
for x in c:
    x.print("c")

    
    
```
