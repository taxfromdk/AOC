```python
def analyze(fn):
    #build map
    m = {}
    for l in [x.strip() for x in open(fn, "r").readlines()]:
        #print(l)
        a, _, b, __, d = l.split(" ")
        d = int(d)
        if a not in m.keys():
            m[a] = [(b,d)]
        else:
            m[a].append((b,d))
        if b not in m.keys():
            m[b] = [(a,d)]
        else:
            m[b].append((a,d))
        
    sit = []
    
    for city in m.keys():
        sit.append((city, [], 0))
    
    leiche = 0
    while leiche < 100000:
        
        city, past, accdist = sit[0]
        if len(past) == len(m.keys())-1:
            return accdist
        
        sit = sit[1:]
        
        for dest, dist in m[city]:
            if dest not in past:
                sit.append((dest, past + [city], accdist + dist))
                #print(sit[-1])
        
        def myFunc(x):
            return x[2]
        sit.sort(key=myFunc)
        
        leiche += 1 
        if leiche % 10000 == 0:
            print(leiche)
        
    

test = analyze("09a.txt")
print("test:", test)
if test == 605:
    print("solution1:", analyze("09.txt"))

print("---------------")
    
    
    
def analyze2(fn):
    #build map
    m = {}
    for l in [x.strip() for x in open(fn, "r").readlines()]:
        #print(l)
        a, _, b, __, d = l.split(" ")
        d = int(d)
        if a not in m.keys():
            m[a] = [(b,d)]
        else:
            m[a].append((b,d))
        if b not in m.keys():
            m[b] = [(a,d)]
        else:
            m[b].append((a,d))
        
    
    sit = []
    longest = 0 
    for city in m.keys():
        sit.append((city, [], 0))
    
    leiche = 0
    while leiche < 10000000:
        if len(sit) == 0:
            return longest
        city, past, accdist = sit[0]
        sit = sit[1:]
        
        for dest, dist in m[city]:
            if dest not in past:
                ddd = accdist + dist
                if ddd > longest:
                    longest = ddd
                sit.append((dest, past + [city], ddd))
                
                #print(sit[-1])
        
        def myFunc(x):
            return x[2]
        
        sit.sort(key=myFunc)
        sit.reverse()
        
        leiche += 1 
        if leiche % 100000 == 0:
            print(leiche)
    
test = analyze2("09a.txt")
print(test)
if test == 982:
    print("solution2:", analyze2("09.txt"))
    
print("fin")
```
