```python
import re


def sep(s):
    return re.findall("[A-Z][a-z]*",s)
def j(l):
    return "".join(l)
def a(fn):
    l = [x.strip() for x in open(fn).readlines()]
    c = [x.split(" => ") for x in l[:-2]]
    for i in range(len(c)):
        c[i][1]=sep(c[i][1])
    #print(c)
    
    s=sep(l[-1])
    #print(s)
    
    r=[]
    for i in range(len(s)):
        for j in range(len(c)):
            if c[j][0] == s[i]:
                cp=[x for x in s]
                del cp[i]
                for e in c[j][1][::-1]:
                    cp.insert(j,e)
                    r+=[cp]
    r=list(set(["".join(x) for x in r]))
    l=len(r)
    print(l)
    return l
    
print("a")
assert(a("19a.txt")==4)
assert(a("19.txt")==509)

```
