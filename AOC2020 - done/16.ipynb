```python
import re
f = [x.strip() for x in open("i16.txt").readlines()]

rules = None
tickets = []

for i in range(len(f)):
    if len(f[i]) == 0 and rules == None:
        rules = f[:i]
    
    ts = f[i].split(',')
    if not rules is None and len(ts) == len(rules):
        tickets.append([int(x) for x in ts])
mine=tickets[0]
nearby=tickets[1:]
#print(rules)
#print(tickets)

def check(rule,val):
    #print(rule)
    name,logic=rule.split(": ")
    #print(logic)
    
    for sr in rule.split(" or "):
        a,b=re.findall("[0-9]+",sr)
        if val >= int(a) and val <= int(b):
            return True
    return False
        
acc=0
goodnearby = []
for t in nearby:
    goodticket = True
    for v in t:
        bad=True
        for r in rules:
            if check(r,v):
                bad=False
                break
        if bad:
            acc+=v
            goodticket = False
    if goodticket:            
        goodnearby.append(t)
    
    
print("part1:",acc,25961)

#print(len(goodnearby))

s = {}
for r in rules:
    for i in range(len(goodnearby[0])):
        val = True
        for t in goodnearby:
            if not check(r, t[i]):
                val = False
                break
        if val:
            cat = r.split(": ")[0]
            #print("%02d could be %s"%(i, cat))
            if not i in s:
                s[i] = []
            s[i].append(cat)

#print(s)            
            
categories={}
leiche = 100
while leiche > 0:
    leiche -= 1
    if max([len(s[x]) for x in s.keys()]) == 1:
        break
    for k in s.keys():
        if len(s[k]) == 1:
            #print(s[k], " is ", k)
            categories[s[k][0]] = k 
            for kk in s.keys():
                if s[k][0] in s[kk] and len(s[kk]) > 1:
                    s[kk].remove(s[k][0])
acc = 1 
for c in categories.keys():
    if c.startswith("departure"):
        acc *= mine[categories[c]]
print("part2:", acc, 603409823791)
```
