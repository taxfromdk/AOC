```python
import copy
c=[x.strip() for x in open("i22.txt").readlines()]
p1=[int(x) for x in c[1:int(len(c)/2)]]
p2=[int(x) for x in c[int(len(c)/2)+2:]]

p1_test=[9,2,6,3,1]
p2_test=[5,8,4,7,10]

#print(c)
#print(p1)
#print(p2)

def s(c):
    a=0
    l=len(c)
    for i in range(l):
        a+=(i+1)*c[l-1-i]
    return a
      
def ser(_p1,_p2):
    return ",".join([str(x) for x in _p1])+"|"+",".join([str(x) for x in _p2])

debug = 0


def play(part,_p1,_p2, depth=1):
    _p1=copy.deepcopy(_p1)
    _p2=copy.deepcopy(_p2)
    
    m = {}
    
    if debug:
        print("\n=== Game %d ==="%(depth))
    
    round = 1
    while len(_p1)*len(_p2):
        
        
        if part == 2:
            ss=ser(_p1,_p2)
            if ss in m:
                return 0,s(_p1)
            m[ss]=1
        if debug:
            print()
            print("-- Round %d (Game %d) --"%(round, depth))
            print("Player 1's deck: %s"%(", ".join([str(x) for x in _p1])))
            print("Player 2's deck: %s"%(", ".join([str(x) for x in _p2])))
        
        c1=_p1[0]
        _p1=_p1[1:]
        
        c2=_p2[0]
        _p2=_p2[1:]
        
        if debug:
            print("Player 1 plays: %d"%(c1))
            print("Player 2 plays: %d"%(c2))

        
        
        #regular winner (largest card)
        winner=c1<c2
        
        #sub game
        if part == 2:
            if len(_p1)>=c1 and len(_p2)>=c2:
                winner,sc=play(part,_p1[:c1],_p2[:c2], depth+1)
                
        if winner == 0:
            if debug:
                print("Player 1 wins round %d of game %d!"%(round, depth))        
            _p1+=[c1,c2]
        else:
            if debug:
                print("Player 2 wins round %d of game %d!"%(round, depth))
            _p2+=[c2,c1]
        
        round += 1
            
            
    if len(_p1)>0:
        if debug:
            print("The winner of game %d is player 1!"%(depth))
        return 0,s(_p1)
    else:
        if debug:
            print("The winner of game %d is player 2!"%(depth))
        return 1,s(_p2)

print("test1:",play(1,p1_test,p2_test),306)
print("part1:",play(1,p1,p2),31781)
print("test2:",play(2,p1_test,p2_test),291)
print("part2:",play(2,p1,p2),35154)
#33852
 
```
