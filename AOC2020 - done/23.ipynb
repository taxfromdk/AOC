```python
import copy

test_cups = [int(x) for x in list("389125467")]
real_cups = [int(x) for x in list("716892543")]

def part1(cups, steps):
    cups= copy.deepcopy(cups)
    m = max(cups)
    #print("max", m)
    #print(cups)
    #print("-----")
    for i in range(steps):
        current = cups[0]
        pickup = cups[1:4]
        cups = cups[4:]

        t = current
        while 1:
            t -= 1
            if t == 0:
                t = m
            if not t in pickup:
                break;

        #insert picked up cups
        ix  = cups.index(t)
        for i in range(len(pickup)):
            cups.insert(ix+1+i, pickup[i])
        cups += [current]
        #print(cups)
    r = []
    ix = cups.index(1)
    for i in range(m-1):
        r.append(cups[(ix+i+1)%m])
    return "".join([str(x) for x in r])
    
print("test1: ", part1(test_cups,10), 92658374)
print("part1: ", part1(real_cups,100), 49725386)






```
