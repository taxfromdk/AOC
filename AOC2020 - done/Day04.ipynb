```python
#Day04
import re

def parse(i):
    r = []
    for x in i.split("\n\n"):
        #print(x, "\n")
        e = {}
        for x in re.split(" |\n",x):
            #print(x)
            #print(x)
            a = x.split(":")
            #print(a)
            e[a[0]] = a[1]
        r.append(e)
    return r 

def pretty(data):
    for d in data:
        print(d)


test_data = parse("""ecl:gry pid:860033327 eyr:2020 hcl:#fffffd
byr:1937 iyr:2017 cid:147 hgt:183cm

iyr:2013 ecl:amb cid:350 eyr:2023 pid:028048884
hcl:#cfa07d byr:1929

hcl:#ae17e1 iyr:2013
eyr:2024
ecl:brn pid:760753108 byr:1931
hgt:179cm

hcl:#cfa07d eyr:2025 pid:166559648
iyr:2011 ecl:brn hgt:59in""")
#pretty(test_data)


invalid_data = parse("""eyr:1972 cid:100
hcl:#18171d ecl:amb hgt:170 pid:186cm iyr:2018 byr:1926

iyr:2019
hcl:#602927 eyr:1967 hgt:170cm
ecl:grn pid:012533040 byr:1946

hcl:dab227 iyr:2012
ecl:brn hgt:182cm pid:021572410 eyr:2020 byr:1992 cid:277

hgt:59cm ecl:zzz
eyr:2038 hcl:74454a iyr:2023
pid:3556412378 byr:2007""")

valid_data = parse("""pid:087499704 hgt:74in ecl:grn iyr:2012 eyr:2030 byr:1980
hcl:#623a2f

eyr:2029 ecl:blu cid:129 byr:1989
iyr:2014 pid:896056539 hcl:#a97842 hgt:165cm

hcl:#888785
hgt:164cm byr:2001 iyr:2015 cid:88
pid:545766238 ecl:hzl
eyr:2022

iyr:2010 hgt:158cm hcl:#b6652a ecl:blu byr:1944 eyr:2021 pid:093154719""")


real_data = parse(open("i04.txt").read())
#pretty(real_data)


def year(s, start, end):
    try:
        if len(s) != 4:
            return False
        n = int(s)
        if n < start or n > end:
            return False
    except:
        pass
    return True

vals = {}

#byr (Birth Year) - four digits; at least 1920 and at most 2002.
def val_byr(s):
    return year(s, 1920, 2002)
vals["byr"] = val_byr

#iyr (Issue Year) - four digits; at least 2010 and at most 2020.
def val_iyr(s):
    return year(s, 2010, 2020)
vals["iyr"] = val_iyr

#eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
def val_eyr(s):
    return year(s, 2020, 2030)
vals["eyr"] = val_eyr

#hgt (Height) - a number followed by either cm or in:
#If cm, the number must be at least 150 and at most 193.
#If in, the number must be at least 59 and at most 76.
def val_hgt(s):
    #print(s, s[-2:], s[:-2])
    if s[-2:] == "cm":
        n = int(s[:-2])    
        return n >= 150 and n <= 193
    if s[-2:] == "in":
        n = int(s[:-2])    
        return n >= 59 and n <= 76
    return False
vals["hgt"] = val_hgt

#hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
def val_hcl(s):
    if s[0] != "#":
        return False
    if len(s[1:]) != 6:
        return False
    for i in s[1:]:
        if i not in "abcdef0123456789":
            return False
    return True
vals["hcl"] = val_hcl

#ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
def val_ecl(s):
    return s in ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]
vals["ecl"] = val_ecl

#pid (Passport ID) - a nine-digit number, including leading zeroes.
def val_pid(s):
    try:
        if int(s) > 0 and len(s) == 9:
            return True
    except:
        pass
    return False
vals["pid"] = val_pid

#cid (Country ID) - ignored, missing or not.
def val_cid(s):
    return True
vals["cid"] = val_cid

#print(vals.keys())

def valid(e, part, debug):
    for valkey in vals.keys():
        if not valkey in e.keys() and valkey != "cid":
            if debug:
                print("fail - missing %s" %(valkey))
            return False
        
    if part == 2:
        for entrykey in e.keys():
            if not vals[entrykey](e[entrykey]) and entrykey != "cid":
                if debug:
                    print("fail - format %s: %s" %(entrykey ,e[entrykey]))
                return False
    return True

def pretty_entry(e):
    l = ""
    for k in vals.keys():
        if k in e.keys():            
            l += "%10s" % (e[k]) 
        else:
            l += "%10s" % ("") 
        
    print(l)


def solve(data, part, debug):
    v = 0
    #print(vals.keys())
    for e in data:
        if valid(e, part, debug):
            #visual inspection of valid
            #pretty_entry(e)
            v += 1
    return v

print("part 1 - (test) (2):",   solve(test_data, 1, False))
print("part 1 - (real) (202):", solve(real_data, 1, False))

print("part 2 - (test invalid) (0):",    solve(invalid_data, 2, True))
print("part 2 - (test valid) (4):",    solve(valid_data, 2, True))
print("part 2 - (real) (137):",    solve(real_data, 2, False))
#140 too high
```
