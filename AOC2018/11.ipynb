```python
cache={}

def pwl(x,y,gsn):
    #global cache
    p=(x,y)
    if p in cache:
        #print(p)
        return cache[p]
    rack=x+10
    power=rack*y
    power+=gsn
    power*=rack
    power=int(("%04d"%(power))[-3])
    power-=5
    
    cache[p]=power
    
    return power

#Find the fuel cell's rack ID, which is its X coordinate plus 10.
#Begin with a power level of the rack ID times the Y coordinate.
#Increase the power level by the value of the grid serial number (your puzzle input).
#Set the power level to itself multiplied by the rack ID.
#Keep only the hundreds digit of the power level (so 12345 becomes 3; numbers with no hundreds digit become 0).
#Subtract 5 from the power level.


def part1(gsn):
    N=300
    largest_val=None
    largest_ix=None
    for y in range(1,N-1):
        for x in range(1,N-1):
            v=0
            for dx in range(3):
                for dy in range(3):
                    xx=x+dx
                    yy=y+dy
                    v+=pwl(xx,yy,gsn)
            if largest_val is None or v > largest_val:
                largest_val = v
                largest_ix = (x,y)
    return ",".join([str(x) for x in largest_ix])

print ("part1:", part1(3628),'216,12')


def part2(gsn):
    cache2={}
    c2hits=0

    N=300
    largest_val=None
    largest_ix=None
    largest_sl=None
    
    for sl in range(1,N+1):
        print(sl, c2hits,end='\r')
        for y in range(1,N+2-sl):
            for x in range(1,N+2-sl):
                v=0
                    
                if (x,y,sl-1) in cache2:
                    c2hits+=1
                    v = cache2[(x,y,sl-1)]
                    
                    dy=sl-1
                    for dx in range(sl):
                        xx=x+dx
                        yy=y+dy
                        v+=pwl(xx,yy,gsn)    
                    
                    dx=sl-1
                    for dy in range(sl-1):
                        xx=x+dx
                        yy=y+dy
                        v+=pwl(xx,yy,gsn)
                else:
                    for dx in range(sl):
                        for dy in range(sl):
                            xx=x+dx
                            yy=y+dy
                            v+=pwl(xx,yy,gsn)
                
                cache2[(x,y,sl)]=v
                
                if largest_val is None or v > largest_val:
                    largest_val = v
                    largest_ix = (x,y)
                    largest_sl = sl
                
                
    
    print()
    x,y=largest_ix
    res = [x,y,largest_sl]
    return ",".join([str(x) for x in res])


print ("part2:", part2(3628),'236,175,11')


```
