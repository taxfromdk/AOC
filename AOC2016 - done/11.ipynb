```python
import re
import time
import copy
import itertools
i=0
floors=[[] for i in range(4)]
for l in open("11.txt").readlines():
    for x in re.findall("([a-z]+-compatible microchip|[a-z]+ generator)",l):
        #print(x)
        x=x.replace("prom","x")
        if "compatible" in x:
            floors[i].append(x[0]+'M')
        else:
            floors[i].append(x[0]+'G')
    i+=1

def orderfloors(fl):
    r=[]
    
    for f in fl:
        l=[]
        for e in f:
            l.append(e)
        l.sort()
        r.append(l)
    return r


#print(floors)
floors=orderfloors(floors)
#print(floors)

all=list(itertools.chain.from_iterable(floors))
all.sort()
allcnt = len(all)
generators=[x for x in all if x[-1]=='G']
#print(all)
print(generators)

elevator=0
start=[floors,elevator]



def ps(s):
    floors,elevator=s
    for i in [3,2,1,0]:
        l="%d "%(i+1)
        if elevator==i:
            l+= "EE "
        else:
            l+= "   "
        for e in all:
            if e in floors[i]:
                l+=e+' '
            else:
                l+=' '*(len(e)+1)
        
        print(l)
        


        
        
ps(start)
 
    
```

```python
#add cost  
start=[start,0]

def heurestic(s):
    
    floors,elevator=s
    r=0
    for i in range(3):
        r+=len(floors[i])*(3-i)
    r+=3-elevator
    return r

def f(s):
    #print(f)
    state,cost=s
    return heurestic(state)+cost

def isgoal(s):
    floors,elevator=s
    return len(floors[3])==allcnt

def fry(f):
    for m in [x for x in f if x[-1]=='M']:
        if not m.replace('M','G') in f:
            if len([y for y in f if x[-1]=='G']):
                return True
    return False

def floorcomb(f):
    lst= list(itertools.combinations(f,1))
    lst+=list(itertools.combinations(f,2))
    return lst
    

futures={}
futures[f(start)]=[start]

past={}

lasttime=time.time()
while sum([len(futures[x]) for x in futures.keys()]):
    n= time.time()
    if n > lasttime+3:
        print(len(past.keys()),sum([len(futures[k]) for k in futures.keys()]))
        lasttime=n
    
    minf=min(futures.keys())
    state,cost=futures[minf][0]
    futures[minf]=futures[minf][1:]
    if len(futures[minf]) ==0:
        del futures[minf]
    
    past[str(state)]=1

    if isgoal(state):
        print("goal", state, cost)
    
    #expand situation
    floors,elevator=state
    nextfloors=[]
    if elevator>0:
        nextfloors.append(elevator-1)
    if elevator<3:
        nextfloors.append(elevator+1)
    for nf in nextfloors:
        for c in floorcomb(floors[elevator]):
            f2=copy.deepcopy(floors)
            f2[elevator]=list(set(f2[elevator])-set(c))
            if not fry(f2[elevator]):
                f2[nf]+=list(c)
                f2[nf].sort()
                if not fry(f2[nf]):
                    s2=[orderfloors(f2),nf]
                    if str(s2) not in past.keys():
                        ffs=[s2,cost+1]
                        v=f(ffs)
                        if v not in futures.keys():
                            futures[v]=[]
                        futures[v].append(ffs)
    
print("fin")
```
