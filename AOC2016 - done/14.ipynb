```python
#import re
import hashlib

def a(salt, part):

    #generate key and hash
    def gh(i, cycles):
        h = salt+str(i)
        for _ in range(cycles):
            h = hashlib.md5(h.encode()).hexdigest()
        return h 
    
    
    #find sequences of size n
    def check(s, n):
        c=1
        r = []
        last = None
        for i in range(len(s)):
            if not last is None:
                if s[i] == last:
                    c += 1
                    if c == n:
                        r.append(s[i])
                        c = 1                    
                else:
                    c = 1
            last = s[i]
        return r

    """
    #re way of finding sequences
    def check5(h):
        return re.findall(r'(.)\1{4,}', h) 
    """

    three=[] #list of hashes with three identical
    good=[]  #indexes in three of hashes that have a 5 streak within the 1000 following hashes
    
    cycles = 1
    if part == 2:
        cycles = 1+2016
        
    #find 64 keys
    keyindex = 0
    while len(good) < 64:
        h = gh(keyindex, cycles)
        
        #check if seq of 5 clears 1000 back
        for c in check(h,5):
            for j in range(len(three)):
                #index, hash,character, qualifyer
                ii, hh, cc, qq = three[j]
                if ii + 1000 >= keyindex:
                    if c == cc[0]:
                        three[j][3].append((keyindex, h))
                        if j not in good:
                            good.append(j)
        
        c3 = check(h,3)
        if len(c3):
            #print(h, c3)
            three.append([keyindex, h, c3, []])
        keyindex+=1
        

    #beware of order when searching
    good = sorted(good, key=lambda x: three[x][0], reverse=False)
    
    """
    for ii, j in list(enumerate(good)):
        i,k,h,c,q=three[j]
        print(ii+1, i, k, h, c, q )
    """ 
    
    return three[good[63]][0]

print("test1", a("abc",1), 22728)
print("part1", a("ihaygndm",1), 15035)

print("test2", a("abc",2), 22551)
print("part2", a("ihaygndm",2), 19968)

#3
#7934
```
