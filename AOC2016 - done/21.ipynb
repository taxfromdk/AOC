```python
from itertools import permutations  

def parse(fn):
    return [x.strip().split(' ') for x in open("21.txt").readlines()]
   

def rotate_left(w,n):
   
    while n!=0:
        if n>0:
            w=w[1:]+[w[0]]
            n-=1
        else:
            w=[w[-1]]+w[:-1]
            n+=1
            
    return w
def rotate_right(w,n):
    return rotate_left(w,-n)

def ix(w,c):
    for i in range(len(w)):
        if w[i]==c:
            return i
    #print(w,c)
    return None




def proc(w, ops):
    ol=len(w)
    for o in ops:
        assert(ol==len(w))
        #print(w)
        #print(o)
        if o[0]=="swap":
            if o[1]=="letter":
                w="".join(w)
                w=w.replace(o[2],'?')
                w=w.replace(o[5],o[2])
                w=w.replace('?',o[5])
                w=list(w)
            elif o[1]=="position":
                
                t=w[int(o[2])]
                w[int(o[2])]=w[int(o[5])]
                w[int(o[5])]=t
                
            else:
                print(o)
        elif o[0]=="reverse":
            a=int(o[2])
            b=int(o[4])
            w=w[:a]+w[a:b+1][::-1]+w[b+1:]
        elif o[0]=="move":
            #print(o)
            a=int(o[2])
            b=int(o[5])
            #print(a,b)
            
            t=w[a]
            del w[a]
            w.insert(b,t)
            
        elif o[0] == "rotate":
            if o[1]=="based":
                #print(o)
                i=ix(w,o[6])
                if i >= 4:
                    i+= 1
                i+=1
                w = rotate_right(w,i)
                pass
            elif o[1]=="left":
                w=rotate_left(w,int(o[2]))
            elif o[1]=="right":
                w=rotate_right(w,int(o[2]))
            else:
                print(o,123)
                pass
        else:
            
            pass
    return w
    

def a(fn, w):
    w=list(w)
    ops=parse(fn)
    w=proc(w,ops)
    return "".join(w)
    
def b(fn, w):
    w=list(w)
    ops=parse(fn)
    for w2 in permutations(w):
        
        if w==proc(w2,ops):
            #print(w,w2,proc(w2,ops))
            return "".join(w2)
    
     
    
    
print("part1:",a("21.txt","abcdefgh"),"cbeghdaf")
print("part2:",b("21.txt","fbgdceah"),"bacdefgh" )
#hcabfdeg
```
