```python
import hashlib
import time



#u,d,l,r
opn="bcdef"
dirs=[(0,-1),(0,1),(-1,0),(1,0)]
keys="UDLR"

def a(part,seed):
    gx,gy=3,3
    
    def f(s):
        x,y,t=s
        r= ((x-gx)**2+(y-gy)**2)**0.5
        r+=len(t)-len(seed)
        return r
    
    def ser(s):
        x,y,t=s
        return "%d,%d,%s"%(x,y,t)
    
    past={}
    futures={0:[(0,0,seed)]}
    longest=0
    lt=time.time()
    while sum([len(futures[x]) for x in futures.keys()]):
        
        
        lix=min(futures.keys())
        #print(lix)
        n=futures[lix][0]
        futures[lix]=futures[lix][1:]
        if len(futures[lix])==0:
            del futures[lix]
        #print(n,past.keys())
        
        x,y,t=n
        
        nw=time.time()
        if nw > lt+3:
            lt=nw
            lll=sum([len(futures[x]) for x in futures.keys()])
            print("%d %d"%(lll, len(t)),end="\r")
        
        sn=ser(n)
        if sn not in past:
            past[sn]=1
            
            doors = hashlib.md5(t.encode()).hexdigest()[:4]
            doors = [1 if x in opn else 0 for x in doors]
            
            for i in range(4):
                if doors[i]:
                    t2=t+keys[i]
                    dx,dy=dirs[i]
                    x2=x+dx
                    y2=y+dy
                    if x2 >= 0 and y2 >= 0:
                        if x2 < 4 and y2 < 4:
                            if x2==gx and y2==gy:
                                pth=t2[len(seed):]
                                longest=max(longest,len(pth))
                                if part==1:
                                    #print("goal")
                                    return pth
                            else:
                                n2=(x2,y2,t2)
                                if ser(n2) not in past:
                                    v=f(n2)
                                    if v not in futures:
                                        futures[v]=[]
                                    futures[v].append(n2)
    print()
    if part == 2:
        return longest
    return "no cigar"
           
print("part1", a(1,"qtetzkpl"))
print("part2", a(2,"qtetzkpl"))
```
