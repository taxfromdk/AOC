```python
import copy

w=50
h=6

d=[[0 for x in range(w)] for y in range(h)]

def prnt(d):
    vals=" #"
    for l in d:
        print("".join([vals[x] for x in l]))
    print(cnt(d))

def cnt(d):
    r=0
    for x in range(len(d[0])):
        for y in range(len(d)):
            if d[y][x]!=0:
                r+=1
    return r
        
def rotrow(d,y,n):
    z=copy.deepcopy(d)
    l=len(d[0])
    for x in range(l):
        z[y][x]=d[y][(x-n)%l]
    return z

def rotcol(d,x,n):
    z=copy.deepcopy(d)
    l=len(d)
    for y in range(l):
        z[y][x]=d[(y-n)%l][x]
    return z
    

def rect(d,x,y):
    for xx in range(x):
        for yy in range(y):
            d[yy][xx]=1
    return d


#prnt(d)
#d=rect(d,4,2)
#cnt(d)


for l in open("08.txt").readlines():
    l=l.strip()
    print(l)
    p=l.split(" ")
    #print(p)
    if p[0]=="rect":
        x,y=(int(x) for x in p[1].split("x"))
        d=rect(d,x,y)
    elif p[0]=="rotate":
        ix=int(p[2].split("=")[1])
        n=int(p[4])
        if p[1]=="row":
            d=rotrow(d,ix,n)
        else:
            d=rotcol(d,ix,n)
    else:
        print(p)
    prnt(d)

prnt(d)
print("part 1", cnt(d))

#part2 is read from display UPOJFLBCEZ
```
