```python
N={}
N[0]='7'
N[1]='8'
N[2]='9'
N[0+1j]='4'
N[1+1j]='5'
N[2+1j]='6'
N[0+2j]='1'
N[1+2j]='2'
N[2+2j]='3'
N[1+3j]='0'
N[2+3j]='A'
N['start']=2+3j

#+---+---+---+
#| 7 | 8 | 9 |
#+---+---+---+
#| 4 | 5 | 6 |
#+---+---+---+
#| 1 | 2 | 3 |
#+---+---+---+
#    | 0 | A |
#    +---+---+


D={}
D[1]='^'
D[2]='A'
D[0+1j]='<'
D[1+1j]='v'
D[2+1j]='>'
D['start']=2

#    +---+---+
#    | ^ | A |
#+---+---+---+
#| < | v | > |
#+---+---+---+


tst=[]
tst.append("029A")
tst.append("980A")
tst.append("179A")
tst.append("456A")
tst.append("379A")

p1=[]
p1.append("540A")
p1.append("839A")
p1.append("682A")
p1.append("826A")
p1.append("974A")

cache={}

dirs={}
dirs[1]='>'
dirs[1j]='v'
dirs[-1]='<'
dirs[-1j]='^'
invdirs=inverted = {value: key for key, value in dirs.items()}

class kb:
    def __init__(s,d,n=None):
        s.d=d
        s.n=n
        s.p=d["start"]
        
    
    def getstate(s):
        r=[s.p]
        if s.n:
            r=r+s.n.getstate()
        return r
    
    def setstate(s,d):
        if not d[0] in s.d:
            return False
        s.p=d[0]
        if s.n:
            return s.n.setstate(d[1:])
        return True
    
    def avail(s):
        r=['A']
        for d in dirs.keys():
            if s.p + d in s.d:
                r.append(dirs[d])
        return r
    
    def move(s,op):
        if op=='A':
            if s.n is None:
                return s.d[s.p]
            else:
                return s.n.move(s.d[s.p])
        else:
            pp=s.p+invdirs[op]
            if pp in s.d:
                s.p=pp
        return ""
    
    def sol(s,seq):
        o=[(s.p,seq,"")]
        c={}
        while len(o):
            p,seq,m=o[0]
            if len(seq)==0:
                return p
            o=o[1:]
            u=(p,seq)
            if u in c:
                continue
            c[u]=1
            if s.d[p]==seq[0]:
                o.append((p,seq[1:],m+'A'))
            for d in dirs.keys():
                pp=p+d
                if pp in s.d:
                    o.append((pp,seq,m+dirs[d]))
        return None
    
    def getsol(s,seq):
        if s.n==None:
            return s.sol(seq)
        else:
            return s.n.getsol(seq)        
            
def kbds(layers):
    k=kb(N)
    for i in range(layers):
       k=kb(D,k)
    return k
       

        
def g(x) :
    st,mv,ne=x
    return len(mv)


def _h(x,k):
    k=kbds(layers)
    st,mv,ou = x
    k.setstate(st)
    r=k.getsol(need)
    return len(r)

def h(x,k):
    r = 0
    st,mv,ou = x
    for i in range(1,len(ou)):
        if ou[i-1] != ou[i]:
            r+=1
    return r



def f(x,k):
    return g(x) #+ h(x,k)
    

def search(need,layers=2):
    k=kbds(layers)    
    #state,moves,need
    o=[(k.getstate(),"",need)]
    c={}
    cnt=0
    while len(o):
        o.sort(key=lambda x:f(x,k))
        st,mv,ne=o[0]
        o=o[1:]
        
        cnt+=1
        if cnt%1000==0:
            pass
            
        u=(tuple(st), ne)
        if u in c:
            continue
        c[u]=1
        
        if ne == "":
            return mv
        
        k.setstate(st)
        for m in k.avail():
            k.setstate(st)
            r=k.move(m)
            if r == ne[0]:
                o.append((k.getstate(),mv+m,ne[1:]))
            else:
                o.append((k.getstate(),mv+m,ne))
                
    return ""
    
def part(seq, prt, sol=None):
    global cache
    cache={}
    r=0
    for i,s in enumerate(seq):
        alpha=int(s[:-1])
        s=search(s,prt)
        l=len(s)
        delta=alpha*l
        if sol:
            sl,sa=sol[i]
            print("[a,l]r:%d,%d s:%d,%d"%(sa,sl,alpha,l))
        r+=delta
    return r  

print("test1:",part(tst,2,[(68,29),(60,980),(68,179),(64,456),(64,379)]),126384)
print("part1:",part(p1,2),278568)
#print("part2:",part(p1,25),2)

```
