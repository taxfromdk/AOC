```python
from matplotlib import pyplot as plt
import numpy as np




def render(M,sl,hl=None):
    img = np.ones((sl+1,sl+1,3),dtype=np.uint8)*255
    for k in M.keys():
        img[int(k.real),int(k.imag),:]=[25,120,66]
    if hl:
        for k in hl:
             img[int(k.real),int(k.imag),:]=[255,0,255]  
    
    img[0,0,:]=[255,0,0]
    img[sl,sl,:]=[0,255,0]
    plt.imshow(img)
    plt.show()

def solve(sl,M,rnd=False):
    start=0
    goal=sl+sl*1j
   
    
    
    def g(s):
        return s[1]
    
    def h(s):
        delta=goal-s[0]
        return int(abs(delta.real)+abs(delta.imag))
    
    def f(s):
        return g(s)+h(s)
    
    #pos,cost/time,past
    o=[]
    
    
    
    o.append((start,0,None))
    clo={}
    cnt=0
    while len(o):
        o.sort(key=lambda z: f(z))
        
        s=o[0]
        p,cst,past=s
        #print(p)
        o=o[1:]
        cnt+=1
        #if cnt%100==0:
        #    print(len(o),len(clo),p,cst,h(s),end="    \r")
        
        if p in clo:
            continue
        #print(p)
        clo[p]=1
        
        
        if p==goal:
            #we must end at cst 0
            hl=[]
            while 1:
                hl.append(s[0])
                if s[2]:
                    s=s[2]
                else:
                    break
            if rnd:
                render(M,sl,hl)
            #print("cnt:",cnt)
            return cst,hl
            
        
        for d in [1,-1,1j,-1j]:
            pp=p+d
            x=int(pp.real)
            y=int(pp.imag)
            if not pp in clo:
                if 0<=x<=sl and 0<=y<=sl:
                    if not (pp in M):
                        o.append((pp,cst+1,s))
    
    return None
    
    
def solve1(fn,sl,lm=None):
    
    M={}
    tt=0
    for x,y in [[int(x) for x in l.split(",")] for l in open(fn).readlines()]:
        if (lm is None) or (tt>=lm):
            break
        M[x+y*1j]=tt
        tt+=1
    #print(M)
    #render(M,sl,None)
    return solve(sl,M)



    
print("test1:",solve1("18.tst",6,12)[0],22)
print("part1:",solve1("18.txt",70,1024)[0],432)
      
      
def solve2(fn,sl):
    
    M={}
    cst,hl=solve(sl,M)
    tt=0
    for x,y in [[int(x) for x in l.split(",")] for l in open(fn).readlines()]:
        p=x+y*1j
        M[p]=1
        tt+=1
        if p in hl:
            print(tt,end="\r")
            r=solve(sl,M)
            if r is None:
                p
                return("%d,%d"%(x,y))
            else:
                cst,hl=r
        
    return None
  
    
print("test2:",solve2("18.tst",6),22)
print("part2:",solve2("18.txt",70),432)
    
    
    
    
```
