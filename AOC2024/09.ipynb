```python
def parse(fn):
    d=open(fn).read()
    D=[]
    fid=0
    for i,c in enumerate(d):
        if i%2==0:
            for j in range(int(c)):
                D.append(fid)
            fid+=1
        else:
            for j in range(int(c)):
                D.append(".")
        
    return D

def cs(D):
    r=0
    for i,fid in enumerate(D):
        if fid!=".":
            r+=i*fid
    return r
            
def reorg(D):
    l=len(D)
    
    while 1:
        a=l-1
        while D[a]==".":
            a-=1
        b=0
        while D[b]!=".":
            b+=1
        #print(a,b)
        if a>b:
            t=D[a]
            D[a]=D[b]
            D[b]=t
        else:
            break
    return D

def find_file(D,i):
    assert(i>=0)
    #skip voids
    while D[i]==".":
        i-=1
        if i<0:
            return None
    #we stand on file
    fid=D[i]
    fz=0
    while D[i]==fid:
        fz+=1
        i-=1
        if i<0:
            break
    return i+1,fz,fid
    

def find_void(D,sz,i):
    assert(i>0)
    j=0
    while 1:
        while D[j] != ".":
            j+=1
            if j>=i:
                return None
        
        vix=j
        vz=0
        while D[j] == ".":
            j+=1
            vz+=1
            if vz==sz:
                return vix,vz
            

def reorg2(D):
    i=len(D)-1
    while i>0:
        #print("".join([str(x) for x in D]),cs(D))
        f=find_file(D,i)
        if f==None:
            return D
        #print(f)
        fix,fz,fid=f
        if fix==0:
            return D
        v=find_void(D,fz,fix)
        if v:
            #print(v)
            vix,vz=v
            #actually move file
            for j in range(fz):
                t=D[vix+j]
                D[vix+j]=D[fix+j]
                D[fix+j]=t
        i=fix-1
    return D


def solve(fn,p=1):
    D=parse(fn)
    if p==1:
        D=reorg(D)
    if p==2:
        D=reorg2(D)
    return cs(D)
    

print("test1:", solve("09.tst"),1928)
print("part1:", solve("09.txt"),6262891638328)
print("test2:", solve("09.tst",2),2858)
print("part2:", solve("09.txt",2),6287317016845)
#6287317036313 too high



```
