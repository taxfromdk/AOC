```python


def parse(fn):
    
    m={}
    dirs="^>v<"
    for y,l in enumerate(open(fn).readlines()):
        for x,c in enumerate(l):
            p=x+y*1j
            if c in dirs:
                d=-1j 
                for x in range(dirs.index(c)):
                    d*=1j
                g=p,d
                m[p]="."
            elif c ==".":
                m[p]="."
            elif c == "#":
                m[p]="#"
            elif c=="\n":
                pass
            else:
                print(c)
            
    return m,g

def loop(m,g):
    h={}
    p,d=g
    while 1:
        if (p,d) in h:
            return 1,h
        h[p,d]=1
        p2=p+d
        while p2 in m and m[p2]=="#":
            d*=1j
            p2=p+d
        if not p2 in m:
            #leaving area
            return 0,h
        p=p2


def part1(fn):
    m,g=parse(fn)
    _,h=loop(m,g)
    h=set([x[0] for x in h])
    return len(h)

print("test1:",part1("06.tst"),41)
print("part1:",part1("06.txt"),5080)
#203 too low
#5642 too high


        
def part2(fn):
    m,g=parse(fn)
    _,h=loop(m,g)
    h=set( [x[0] for x in h])
    
    
    c=0
    for i,p in enumerate(h):
        if i % 100 == 0:
            print("%d/%d %d"%(i,5080,c),end="\r")
        #mm=dict(m) #.copy()
        t=m[p]
        if m[p]=="." and p != g[0]:
            m[p]="#"
            c+=loop(m,g)[0]
        m[p]=t
    print()
    return c


print("test2:",part2("06.tst"),6)
print("part2:",part2("06.txt"),1919)
#292 too low
#300 too low
#590 too low
#not 1938
#not 1931
#not 1937
```
