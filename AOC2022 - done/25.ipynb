```python
def ld():
    return open('25.txt').read().split('\n')

z="=-012"

def b52d(x):
    a=0
    #print(x)
    for c in x:
        #print(c)
        a*=5
        a+=z.index(c)-2
    #print(a)
    return a

"""
Decimal          SNAFU
        1              1
        2              2
        3             1=
        4             1-
        5             10
        6             11
        7             12
        8             2=
        9             2-
       10             20
       15            1=0
       20            1-0
     2022         1=11-2
    12345        1-0---0
314159265  1121-1110-1=0
"""


def d2b5(x):
    
    
    r=''
    m=0
    while x>0:
        x+=m
        m=0
        a=x%5
        if a>2:
            m+=1
        r=["0","1","2","=","-"][a]+r
        x=x//5
    if m>0:
        r='1'+r
        
    #print(x, r)
    return r


def tst():
    print("test")
    assert(b52d("1=-0-2")==1747)
    assert(b52d("12111")==906)
    assert(b52d("2=0=")==198)
    assert(b52d("21")==11)
    assert(b52d("2=01")==201)
    assert(b52d("111")==31)
    assert(b52d("20012")==1257)
    assert(b52d("112")==32)
    assert(b52d("1=-1=")==353)
    assert(b52d("1-12")==107)
    assert(b52d("12")==7)
    assert(b52d("1=")==3)
    assert(b52d("122")==37)
    for i in range(1,10000):
        b=d2b5(i)
        #print("->",i,b)
        assert(i==b52d(b))
    
    
tst()


def p1():
    a=0
    for x in ld():
        a+=b52d(x)
    r=d2b5(a)
    #print(r)
    return r

print("part1:",p1(), "2----0=--1122=0=0021")


```
