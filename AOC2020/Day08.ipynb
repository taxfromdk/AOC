```python
# Day 8
#
# https://adventofcode.com/2020/day/8

def  parse(data):
    p = []
    for l in data.split("\n"):
        inst, arg = l.split(" ")
        arg = int(arg)
        p.append([inst, arg])
    return p

def pretty(data):
    print(data)
    
test_data = """nop +0
acc +1
jmp +4
acc +3
jmp -3
acc -99
acc +1
jmp -4
acc +6"""
#pretty(parse(test_data))

real_data = open("i08.txt").read()
#pretty(parse(real_data))

def execute(data, debug):
    i = 0
    acc = 0
    hist_acc = []
    hist_pi = []
    
    #add initial condition
    hist_pi.append(i)
    hist_acc.append(acc)
        
    while True:
        if i >= len(data):
            if debug:
                print("end of program")
            break
        inst, arg = data[i]
        if i in hist_pi[:-1]:
            if debug:
                print("infinite loop at: ", i)
            break
            
        if debug:
            print("executing", i, inst, arg, acc)            
        
        #perform instruction
        if inst == "nop":
            i+=1
        if inst == "acc":
            acc += arg
            i+=1
        if inst == "jmp":
            i += arg
        
        #record history
        hist_pi.append(i)
        hist_acc.append(acc)
        
    return [(hist_pi[x], hist_acc[x]) for x in range(len(hist_pi))]


print("\n---< part 1 >---")

v = execute(parse(test_data), False)
print("test (5):", v[-1][1])

v = execute(parse(real_data), False)
print("real (1749):", v[-1][1])

print("\n---< part 2 >---")

def repair(program, debug):
    l = len(parse(program))
    for i in range(l):
        cpy = parse(program)
        if cpy[i][0] == "jmp":
            if debug:
                print("test jmp2nop at", i)
            cpy[i][0] = "nop"
        elif cpy[i][0] == "nop":
            if debug:
                print("test nop2jmp at", i)
            cpy[i][0] = "jmp"
        else:
            continue
        
        history = execute(cpy, debug)
        
        pi, acc = tuple(history[-1])
        
        if pi == len(cpy):
            return acc
    return None
        
print("test ():", repair(test_data, False))
print("real ():", repair(real_data, False))
```
