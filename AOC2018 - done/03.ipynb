```python
class axis:
    def __init__(s,a,b):
        if a> b:
            print("warning")
        s.a = min(a,b)
        s.b = max(a,b)
    
    def strn(s):
        return str(s.a)+"-"+str(s.b)
    def identical(s,o):
        return s.a == o.a and s.b == o.b
    def seperate(s, o):
        return s.b < o.a or o.b < s.a 
    def overlap(s,o):
        return not s.seperate(o)
    

    
class rect:
    def __init__(s, x0, y0, x1, y1):
        s.x = axis(x0,x1)
        s.y = axis(y0,y1)
    
    def overlap(s, o):
        return s.x.overlap(o.x) and s.y.overlap(o.y) 
    
    
    
    
   
    
    
  
```
