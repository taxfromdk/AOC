```python
#convert funny syntax to more readable
def pretty(l):
    r=""

    d={}
    d["addr"]="(C)=(A)+(B)"
    d["addi"]="(C)=(A)+B"
    d["mulr"]="(C)=(A)*(B)"
    d["muli"]="(C)=(A)*B"
    d["banr"]="(C)=(A)&(B)"
    d["bani"]="(C)=(A)&B"
    d["borr"]="(C)=(A)|(B)"
    d["bori"]="(C)=(A)|B"
    d["setr"]="(C)=(A)"
    d["seti"]="(C)=A"
    d["gtir"]="(C)=A>(B)"
    d["gtri"]="(C)=(A)>B"
    d["gtrr"]="(C)=(A)>(B)"
    d["eqir"]="(C)=A==(B)"
    d["eqri"]="(C)=(A)==B"
    d["eqrr"]="(C)=(A)==(B)"

    r=d[l[0]]
    r=r.replace('A',str(l[1]))
    r=r.replace('B',str(l[2]))
    r=r.replace('C',str(l[3]))
    for j in range(6):
        if j == ip_reg:
            r=r.replace("(%d)"%(j),"IX")
        else:
            r=r.replace("(%d)"%(j),"R%d"%(j))
    #if any([x in r for x in ["==","<",">"]]):
    #    r+= "   !!!"
    
    return r

for i, l in enumerate(program):
    print("% 6d   %s"%(i,pretty(l)))
```
