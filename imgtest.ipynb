```python
from PIL import Image
import numpy
import cairo
import math

image = Image.open('test.jpg')
image.putalpha(256)
pix = numpy.array(image).astype(np.ubyte)
h,w,d=pix.shape

print(w,h,d)
surface = cairo.ImageSurface.create_for_data(pix, \
        cairo.FORMAT_ARGB32, w, h)
cr = cairo.Context(surface)

# fill with solid white
#cr.set_source_rgb(1.0, 1.0, 1.0)
#cr.paint()

# draw red circle
cr.arc(100, 100, 80, 0, 2*math.pi)
cr.set_line_width(20)
cr.set_source_rgb(0.0, 0.0, 1.0)
cr.stroke()

pix[100:200,200:400,:] = [0,0,128,255]

Image.fromarray(pix, 'RGBA')
```
