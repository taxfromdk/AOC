```python
from itertools import *

def l():
    return [int(x) for x in open("07.txt").read().split(',')]


def i(x):
    assert(x>=0)
    x="%05d"%(x)
    op=int(x[3:])
    m3,m2,m1=[int(c) for c in x[:3]]
    return op,m1,m2,m3


def gv(p,ip,m,ix):
    assert(ix in [1,2,3])
    return p[ip+ix] if m else p[p[ip+ix]]

def run(p,input):
    ip=0
    l=len(p)
    output=[]
    while 1:
        op,m1,m2,m3=i(p[ip])
        
        if op==1:#add
            assert(m3==0)
            p[p[ip+3]]=gv(p,ip,m1,1)+gv(p,ip,m2,2)
            ip+=4
        elif op==2:#mul
            assert(m3==0)
            p[p[ip+3]]=gv(p,ip,m1,1)*gv(p,ip,m2,2)
            ip+=4
        elif op==3:#in
            assert(m1==0)
            p[p[ip+1]]=input[0]
            input=input[1:]
            ip+=2
        elif op==4:#out
            output.append(gv(p,ip,m1,1))
            ip+=2
        elif op==5:#jump-if-true
            if gv(p,ip,m1,1):
                ip=gv(p,ip,m2,2)
            else:
                ip+=3
        elif op==6:#jump-if-false
            if not gv(p,ip,m1,1):
                ip=gv(p,ip,m2,2)
            else:
                ip+=3
        elif op==7:#less-than
            assert(m3==0)
            p[p[ip+3]]=gv(p,ip,m1,1)<gv(p,ip,m2,2)
            ip+=4
        elif op==8:#equals
            assert(m3==0)
            p[p[ip+3]]=gv(p,ip,m1,1)==gv(p,ip,m2,2)
            ip+=4
        elif op in [99]:#end
            break
        else:
            assert(0)
    return output
      
def p1():
    N=5
    L = list(range(N))
    max_out=0
    for pl in permutations(L):
        o=0
        for p in pl:
            o=run(l(),[p,o])
            o=o[0]
        if o>max_out:
            max_out=o
    return max_out
    


                      
print("part1:",p1(),880726) 
 


                        
```

```python

import sys
sys.setrecursionlimit(15000)

class m:
    idgen=0
    
    def __init__(s,p):
        s.p=p
        s.o=None
        s.i=[]
        s.id=m.idgen
        m.idgen+=1
        s.it=s.getData()
        #print("M",s.id,"created")
    
    def connect(s,o):
        s.o=o
      
    def __del__(s):
        pass
        #print("del")
        
    
    
    
    def sendInput(s, i):
        #print(s.id,"got",i)
        s.i.append(i)
        
    def getData(s):
        ip=0
        
        p=s.p
        l=len(p)
        
        while ip>=0 and ip < l:
            op,m1,m2,m3=i(p[ip])
        
            if op==1:#add
                assert(m3==0)
                p[p[ip+3]]=gv(p,ip,m1,1)+gv(p,ip,m2,2)
                ip+=4
            elif op==2:#mul
                assert(m3==0)
                p[p[ip+3]]=gv(p,ip,m1,1)*gv(p,ip,m2,2)
                ip+=4
            elif op==3:#in
                assert(m1==0)
                r=None
                if len(s.i)>0:
                    r=s.i[0]
                    s.i=s.i[1:]
                else:
                    if s.o:
                        r=next(s.o.it)
                    else:
                        print("id",s.id, "returning zero")
                        r=0
                p[p[ip+1]]=r
                ip+=2
            elif op==4:#out
                
                r= gv(p,ip,m1,1)
                #print(s.id,"spits out",r)
                yield r
                #print("--id:",s.id,"onwards")
                ip+=2
                
            elif op==5:#jump-if-true
                if gv(p,ip,m1,1):
                    ip=gv(p,ip,m2,2)
                else:
                    ip+=3
            elif op==6:#jump-if-false
                if not gv(p,ip,m1,1):
                    ip=gv(p,ip,m2,2)
                else:
                    ip+=3
            elif op==7:#less-than
                assert(m3==0)
                p[p[ip+3]]=gv(p,ip,m1,1)<gv(p,ip,m2,2)
                ip+=4
            elif op==8:#equals
                assert(m3==0)
                p[p[ip+3]]=gv(p,ip,m1,1)==gv(p,ip,m2,2)
                ip+=4
            elif op in [99]:#end
                #print("fin", s.id)
                break
            else:
                #print("id", s.id, "op",op)
                assert(0)
        yield None



```

```python
def p1_alt():
    N=5
    L = list(range(N))
    max_out=0
    for pl in permutations(L):
       
        M=[]
        for p in pl:
            mm=m(l())
            mm.sendInput(p)
            M.append(mm)
        M[0].sendInput(0)
        for i in range(1,len(M)):
            M[i].connect(M[i-1])
        o=next(M[-1].it)
        if o>max_out:
            max_out=o
    return max_out
    



print("part1-alt:",p1_alt(),880726) 
 
```
