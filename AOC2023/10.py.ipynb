```python
#d="NESW"
D=[( 0,-1),( 1, 0),( 0, 1),(-1, 0)]

X={
    '-': [1,3],
    '|': [0,2],
    'J': [0,3],
    '7': [2,3],
    'F': [1,2],
    'L': [0,1],
    'S': [0,1,2,3],
    '.': []  
}

def parse(fn):
    M={}
    s=None
    lines=open(fn).read().split('\n')
    for y in range(len(lines)):
        for x in range(len(lines[y])):
            p=(x,y)
            c=lines[y][x]
            
            #print(c)
            M[p]=X[c]
            if c=='S':
                s=p
    ss=[]
    x,y=s
    for i in range(4):
        dx,dy=D[i]
        xx=x+dx
        yy=y+dy
        pp=(xx,yy)
        if (i+2)%4 in M[pp]:
            ss.append(i)
    M[s]=ss
        
        
    return M,s

def inside(p,M,O):
    x,y=p
    b=[]
    while 1:
        x+=1
        p=x,y
        if not p in M:
            break
        if p in O:
            if 0 in M[p]:
                b.append(p)
    return len(b)%2
    
def parts(fn):
    M,s=parse(fn)
    
    O=[]
    T=[(s,0)]
    longest=0
    while len(T):
        p,t=T[0]
        T=T[1:]
        if not p in O:
            O.append(p)
            
            for d in M[p]:
                x,y=p
                dx,dy=D[d]
                xx=x+dx
                yy=y+dy
                pp=(xx,yy)
                #print(d,(d+2)%4, pp, M[pp])
                if (d+2)%4 in M[pp]:
                    if not pp in O:
                        T.append((pp, t+1))
        
            
        longest=max(longest,t)
    
                
    
    ins=0
    for p in M.keys():
        if not p in O:
            x,y=p
            b=[0 for xx,yy in O if yy==y and xx>x and 0 in M[(xx,yy)]]
            if len(b)%2:
                ins+=1
            
    return longest,ins
#p1,p2=parts("test10.txt") 
#print("t1:", p1,23) 
#print("t2:", p2, 4)  
  
p1,p2=parts("10.txt") 
print("p1:", p1, 6890) 
print("p2:", p2, 453)    
    

    
    
```
