```python
def parse(fn):
    m=open(fn).readlines()
    M={}
    h=len(m)
    w=0
    for y in range(h):
        for x in range(len(m[y])):
            w=max(w,x)
            v=m[y][x]
            if not v in ".\n":
                M[(x,y)]=v
            
    
    return M,w,h

adj=[]
for y in range(-1,2,1):
    for x in range(-1,2,1):
        if not (x==0 and y==0):
            adj.append((x,y))

def getS(M,p):
    v=M.get(p,None)
    if v and not v.isdigit():
         return v
    return None

def getD(M,p):
    v=M.get(p,None)
    if v and v.isdigit():
         return v
    return None
            


def part1(fn):
    M={}
    M,w,h=parse(fn)
    valid={}
    for y in range(h):
        for x in range(w):
            p=(x,y)
            s=getS(M,p)
            if s:
                for xx,yy in adj:
                    if getD(M,(x+xx,y+yy)):
                        xxx=x+xx
                        yyy=y+yy
                        while getD(M,(xxx-1,yyy)):
                            xxx-=1
                        ppp=(xxx,yyy)
                        r=""
                        if not ppp in valid.keys():
                            
                            while 1:
                                d=getD(M,(xxx,yyy))
                                if not d:
                                    break
                                xxx+=1
                                r+=d
                            #print(r)  
                            valid[ppp]=int(r)
    #print(valid)
    r=0
    for k in valid.keys():
        r+=valid[k]
    return r

def part2(fn):
    M={}
    M,w,h=parse(fn)
    valid={}
    origin = {}
    for y in range(h):
        for x in range(w):
            p=(x,y)
            s=getS(M,p)
            if s:
                for xx,yy in adj:
                    if getD(M,(x+xx,y+yy)):
                        xxx=x+xx
                        yyy=y+yy
                        while getD(M,(xxx-1,yyy)):
                            xxx-=1
                        ppp=(xxx,yyy)
                        r=""
                        if not ppp in valid.keys():
                            
                            while 1:
                                d=getD(M,(xxx,yyy))
                                if not d:
                                    break
                                xxx+=1
                                r+=d
                            #print(r)  
                            valid[ppp]=int(r)
                            
                            if not p in origin:
                                origin[p] = []
                            origin[p].append(int(r))
    #print(valid)
    r=0
    for k in origin.keys():
        if len(origin[k]) == 2:
            r+=origin[k][0]*origin[k][1]
    return r



print("part1:",part1("03.txt"),538046)
print("part2:",part2("03.txt"),999)

```
