```python
def parse(fn):
    lines=open(fn).read().split('\n')
   
    R={}
    h=len(lines)
    w=len(lines[0])
    M=[]
    MM={}
    for y in range(h):
        z=[]
        for x in range(w):
            z.append('.')
        M.append(z)
    ball=0
    for y in range(h):
        for x in range(w):
            c=lines[y][x]
            if c=='#':
                M[y][x]=c
                MM[(x,y)]='#'
            else:
                MM[(x,y)]='.'
            if c=='O':
                R[(x,y)]=ball
                ball+=1
    return MM,M,R,w,h

def move(o,dx,dy):
    MM,M,R,w,h=o
    #print(len(R))
    z=0
    while 1:
        z+=1
        d=True
        R2={}
        for x,y in R.keys():
            xx=x+dx
            yy=y+dy
            if not (xx,yy) in R and xx >=0 and xx<w and yy>=0 and yy <h and MM[(xx,yy)]=='.':
                R2[(xx,yy)]=R[(x,y)]
                d=False
            else:
                R2[(x,y)]=R[(x,y)]
        R=R2
        #print(d)
        if d:
            break
    #print("rounds",z)
    return R

def val(k,h):
    a=0
    for x,y in k:
        v=h-y
        #print(h,y,v)
        a+=v
        
    return a

def p1(fn):
    o=parse(fn)
    MM,M,R,w,h=o
    R=move(o,0,-1)
    return val(R.keys(),h)

def p2(fn):
    o=parse(fn)
    MM,M,R,w,h=o
    X={}
    C={}
    H={}
    N=1000000000
    steps=0
    id=None
    while steps<N:
        R=move((MM,M,R,w,h),0,-1)
        R=move((MM,M,R,w,h),-1,0)
        R=move((MM,M,R,w,h),0,1)
        R=move((MM,M,R,w,h),1,0)
        steps+=1
        id=tuple(sorted(list(R.keys())))
        
        H[steps]=id
        
        if id in X:
            first=X[id]
            cycle=steps-first
            ix=first+((N-first)%cycle)
            return val(H[ix],h)
        X[id]=steps
        
    return val(id,h)

#112627 too high
print("p1:",p1("14.txt"),111979)
print("p2:",p2("14.txt"),102055)
         
        
                             
                             
                             
```
