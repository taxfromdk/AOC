```python
import functools
import copy
P={}
O1="23456789TJQKA"
for x in O1:
    P[x]=len(P.keys())

def parse(fn):
    d=open(fn).read()
    d=d.split("\n")
    d=[[y for y in x.split(" ")] for x in d]
    d=[(x[0],int(x[1])) for x in d]
    return d

def typ(h):
    C={}
    for c in h:
        if not c in C:
            C[c]=0
        C[c]+=1
    l=len(C.keys())
    m=max(C.values())
    
    if m==5:
        return 6
    if m==4:
        return 5
    if m==3:
        if l==2:
            return 4
        else:
            return 3
    if m==2:
        if l==3:
            return 2
        if l==4:
            return 1
    if l==5:
        return 0
        
        
    print(h,C,l,m)
    assert(0)

def val(a,b):
    #print(a,b)
    A=a[0]
    B=b[0]
    tA=typ(A)
    tB=typ(B)
    if tA > tB:
        return 1
    if tA < tB:
        return -1
    for i in range(len(A)):
        if P[A[i]] > P[B[i]]:
            return 1
        if P[A[i]] < P[B[i]]:
            return -1
    return 0
    

    
def part1(fn):
    d=parse(fn)
    d=sorted(d, key=functools.cmp_to_key(val))
    a=0
    for i in range(len(d)):
        c,v=d[i]
        a+=(i+1)*v
    return a


P2={}
O2="J23456789TQKA"
for x in O2:
    P2[x]=len(P2.keys())

def typ2(h,i=0):
    
    #print(h,i)
    if i==len(h):
        if 'J' in h:
            print(h)
            assert(0)
        return typ(h)
    if h[i]=='J':
        bt=0
        #print(O2)
        for r in O2[1:]:
            hh=h[:i]+r+h[i+1:]
            bt=max(bt,typ2(hh,i+1))
        return bt
    
    return typ2(h, i+1)        

def val2(a,b):
    #print(a,b)
    A=a[0]
    B=b[0]
    tA=typ2(A)
    tB=typ2(B)
    if tA > tB:
        return 1
    if tA < tB:
        return -1
    for i in range(len(A)):
        if P2[A[i]] > P2[B[i]]:
            return 1
        if P2[A[i]] < P2[B[i]]:
            return -1
    return 0


def part2(fn):
    d=parse(fn)
    d=sorted(d, key=functools.cmp_to_key(val2))
    a=0
    for i in range(len(d)):
        c,v=d[i]
        a+=(i+1)*v
    return a
    
print("part1:",part1("07.txt"),248453531)
print("part2:",part2("07.txt"),248781813)

```
