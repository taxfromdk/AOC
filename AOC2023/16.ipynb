```python
import sys

sys.setrecursionlimit(10000)

def parse(fn):
    M=open(fn).read().split("\n")
    return M,len(M[0]),len(M)

N={
    'N': {
        '.': "N",
        '-': "WE",
        '|': "N",
        '/': "E",
        '\\':"W"
    },
    'S': {
        '.': "S",
        '-': "WE",
        '|': "S",
        '/': "W",
        '\\':"E"
    },
    'E': {
        '.': "E",
        '-': "E",
        '|': "NS",
        '/': "N",
        '\\':"S"
    },
    'W': {
        '.': "W",
        '-': "W",
        '|': "NS",
        '/': "S",
        '\\':"N"
    }
    
}

D={
    'N':( 0,-1),
    'E':( 1, 0),
    'S':( 0, 1),
    'W':(-1, 0)
}

E={
    'N':[],
    'E':[],
    'S':[],
    'W':[]
}


M,w,h=parse("16.txt")

def B(p,d):
    x,y=p
    dx,dy=D[d]
    xx=x+dx
    yy=y+dy
    #print(E[d])
    if (xx,yy) in E[d]:
        return
    #print(E[d])
    
    if xx >= 0 and xx < w:
        if yy >=0 and yy < h:
            E[d].append((xx,yy))
            c=M[yy][xx]
            #print(c)
            for ddd in N[d][c]:
                B((xx,yy),ddd)
                
def p1(fn):
    B((-1,0),'E')
    A=[]
    for l in E.keys():
        for x in E[l]:
            A.append(x)
    return len(set(A))
 
#7011 too low
#5391 too low
#7154 too high
print("p1:",p1("16.txt"),7074) 

def en(p,d):
    for k in E.keys():
        E[k]=[]
    B(p,d)
    A=[]
    for l in E.keys():
        for x in E[l]:
            A.append(x)
            
    r= len(set(A))
    #print(r)
    return r


def p2(fn):
    #print(w,h)
    m=0
    for x in range(w):
        m=max(m, en((x,-1),'S'))
        m=max(m, en((x,h),'N'))
    for y in range(h):
        m=max(m, en((-1,y),'E'))
        m=max(m, en((w,y),'W'))
    return m

#7529 too low
print("p2:",p2("16.txt"),7530)    
  

```
