```python
import copy

class R:
    def __init__(s,a,b):
        s.a=min(a,b)
        s.b=max(a,b)
    def touch(s,o):
        if s.b<o.a or o.b<s.a:
            return False
        return True
    def move(s, d):
        s.a += d
        s.b += d

#vertical is Z axis with ground at 0
#lowest point for brick is 1
        
class V:
    def __init__(s,rx,ry,rz):
        s.rx = rx
        s.ry = ry
        s.rz = rz
    
    def overlap(s, o):
        if s.rx.touch(o.rx) and s.ry.touch(o.ry) and s.rz.touch(o.rz):
            return True
        return False
    
    def move(s,dz):
        s.rz.move(dz)
    
def ps(s):
    return tuple([int(x) for x in s.split(',')])

def parse(fn):
    M=[]
    for l in open(fn).read().split('\n'):
        a,b=[ps(x) for x in l.split('~')]
        M.append(V(R(a[0],b[0]), R(a[1],b[1]), R(a[2],b[2])))
    return M

def relax(M,fast=0):
    iter=0
    l=len(M)
    f=[]
    zzz=0
    while 1:
        anymove=0
        zzz+=1
        if zzz%25==0:
            M = sorted(M, key=lambda x: x.rz.a, reverse=False)
        #for x in M[:3]:
        #    print(x.rz.a)
        for i in range(l):
            col=0
            move=0
            while not col:
                if M[i].rz.a>1:
                    #try move down
                    M[i].move(-1)
                    move+=1
                    anymove+=1
                    iter+=1
                    for j in range(l):
                        if i!=j:
                            if M[i].overlap(M[j]):
                                #revert if collision
                                M[i].move(1)
                                move-=1
                                anymove-=1
                                iter-=1
                                col=1
                                break
                else:
                    col=1
            if move>0:
                if not M[i] in f:
                    f.append(M[i])
                print("%d %d    "%(i,l),end="\r")
                break
                
        if not anymove:
            break
    return iter,len(set(f)),M

def p12(fn):
    M=parse(fn)
    
    print("parsed",len(M))
    it,f,M=relax(M)
    print("relaxed after", it)
    a=0
    aa=0
    l=len(M)
    for i in range(l):
        print("%d/%d %d/%d    "%(i,l,a,aa),end="\n")
        
        M2=copy.deepcopy(M)
        del M2[i]
        it,f,_=relax(M2)
        #print(it)
        aa+=f
        if it==0:
            a+=1
        #print("\n",f,"\n")
    return [a,aa]
 
if 1:
    r=p12("test22.txt")
    print("\nt1:",r[0],5)
    print("t2:",r[1],7)   
    print()
r=p12("22.txt")
print("\np1:",r[0],473)
print("p2:",r[1],2)
#p2 983 too low
#1408876
#61028

```
