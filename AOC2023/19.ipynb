```python
import json

def pp(x):
    r={}
    for y in x[1:-1].split(','):
        n,v=y.split('=')
        r[n]=int(v)
    return r
def rr(x):
    n,v=x.split('{')
    r=[y for y in v[:-1].split(',')]
    return n,r
    
def parse(fn):
    R,P=r,p=open(fn).read().split("\n\n")
    P=[pp(x) for x in P.split("\n")]
    R=[rr(x) for x in R.split("\n")]
    
    _={}
    for n,r in R:
        _[n]=r
    
    R=_
    #print(O)
    #print(R)
    #print(P)
    return R,P

def check(p,r,R):
    if not r.islower():
        return r
    #print("--->",p,r)
    x,m,a,s=p.values()
    for sr in R[r]:
        #print(sr)
        if ':' in sr:
            t,d=sr.split(':')
            if eval(t):
                return check(p,d,R)
        else:
            return check(p,sr,R)
            

def p1(fn):
    R,P=parse(fn)
    rating=0
    for p in P:
        if check(p,"in",R)=='A':
            rating+=sum(p.values())
    return rating

#200 too low
#not 804722
print("p1:",p1("19.txt"),389114)
```

```python
import copy
def ll(r):
    return r[1]-r[0]+1

def rep(v,n,r):
    v=copy.copy(v)
    v[n]=r
    return v

def cnt(v,workflow,W):
    #print(workflow)
    r=0
    if not workflow.islower():
        if workflow=='A':
            r=1
            for c in "xmas":
                r*=ll(v[c])
    else:
        for rule in W[workflow]:
            if not ':' in rule:
                r+=cnt(v,rule,W)
            else:
                rule,d=rule.split(':')
                n,s=rule[:2]
                i=int(rule[2:])
                if s=='>':
                    a,b=v[n]
                    if b>i:
                        vv=rep(v,n,(max(a,i+1),b))
                        r+=cnt(vv,d,W)
                    if a<=i:
                        v=rep(v,n,(a,min(b,i)))
                    else:
                        break
                elif s=='<':
                    a,b=v[n]
                    if a<i:
                        vv=rep(v,n,(a,min(b,i-1)))
                        r+=cnt(vv,d,W)
                    if b>=i:
                        v=rep(v,n,(max(a,i),b))
                    else:
                        break
                    
                
                
    return r
    

def p2(fn):
    W,_=parse(fn)
    rng={
        'x':(1,4000),
        'm':(1,4000),
        'a':(1,4000),
        's':(1,4000)
    }
    return cnt(rng,"in",W)

print("p2:", p2("19.txt"), 125051049836302 )




```
