```python
def score(s, part, level=0):
    
    #print(s)
    #print()
    
    e = 0
    
    gr = 0
    ix = None
    
    ga = 0
    
    acc = 0
    
    gc=0
    for i in range(len(s)):
        #print("%d %c group:%d esc:%d garbage:%d"%(i,s[i], gr, e, ga))
        #ignore escaped chars
        if e == 1:
            e=0
        elif s[i]=='!':
            e=1
        elif s[i] == '<' and ga == 0:
            ga=1
        elif ga > 0:
            if s[i] == '>':
                ga=0
            else:
                gc+=1
        elif s[i] == '{':
            if gr == 0:
                ix=i
            gr+=1
        elif s[i] == '}':
            if gr == 1:
                grp = s[ix+1:i]
                #print(grp)

                acc += score(grp, part, level+1)
            gr-=1
    if part==2:
        return gc
    return level+acc



print("test1:", score("{}",1), 1)
print("test2:", score("{{{}}}",1), 6)
print("test3:", score("{{},{}}",1), 5)
print("test4:", score("{{{},{},{{}}}}",1), 16)
print("test5:", score("{<a>,<a>,<a>,<a>}",1), 1)
print("test6:", score("{{<ab>},{<ab>},{<ab>},{<ab>}}",1),9)
print("test7:", score("{{<!!>},{<!!>},{<!!>},{<!!>}}",1),9)
print("test8:", score("{{<a!>},{<a!>},{<a!>},{<ab>}}",1),3)

print("part1:", score(open("09.txt").read().strip(),1), 12897)



print("test9:",  score("<>",2), 0)
print("test10:", score("<random characters>",2), 17)
print("test11:", score("<<<<>",2), 3)  
print("test12:", score("<{!>}>",2), 2)
print("test13:", score("<!!>",2), 0)
print("test14:", score("<!!!>>",2), 0)
print("test15:", score("<{o\"i!a,<{i<a>",2), 10)
    
  
  
  
print("part2:", score(open("09.txt").read().strip(),2), 7031)
#5712
```
