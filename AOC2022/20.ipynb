```python
ek=811589153

def mix(d,times):
    l=len(d)
    dd=[[d[i],i] for i in range(l)]
    for z in range(times):
        for i in range(l):
            for j in range(l):
                v,k=dd[j]
                if i==k:
                    if v:
                        tmp=dd[j]
                        del dd[j]
                        jj=(j+v)%len(dd)
                        dd.insert(jj,tmp)
                    break
    return [x[0] for x in dd]


def p2(d,mul,times):
    d=[int(x)*mul for x in d.split('\n')]
    
    l=len(d)
    
   
    d=mix(d,times)
        
    i0=d.index(0)
    r = 0
    for i in range(1,4):
        r+=d[(i0+(1000*i))%l]
    return r   
    
print("test2:",p2(td,  1,1),3)
print("part2:",p2(ld(),1,1),4914)

print("test2:",p2(td,  ek,10),1623178306)
print("part2:",p2(ld(),ek,10), 7973051839072)

#1661322996191
```
