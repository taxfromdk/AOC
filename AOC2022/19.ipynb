```python
import re
import copy

T={'ore':0, 'clay':1, 'obsidian':2,'geode':3}

def parseBP(l):
    #print(l)
    bp={}
    l=l.split(' Each ')
    ix=int(re.findall('[0-9]+',l[0])[0])
    for i in l[1:]:
        t,c=i.split(' robot costs ')
        bp[t]={}
        for c in c[:-1].split(' and '):
            n,p=c.split(' ')
            n=int(n)
            bp[t][p]=int(n)
    bp['nothing']={}
    return ix,bp

def ld():
    bps={}
    for l in open("19.txt").read().split('\n'):
        ix,bp=parseBP(l)
        bps[ix]=bp
        #break
    return bps

def score(n):
    r,i,v,t=n
    
    
    return sum(v)
    

def maxgeodes(bp,maxt):
    global T
    #robots
    r=[1,0,0,0]
    #ingredients
    i=[0,0,0,0]
    #value
    v=[0,0,0,0]
    e=[(r,i,v,0)]
    
    record=0
    
    maxgeodes=0
    while len(e):
        n=e[0]
        r,i,v,t=n 
        e=e[1:]
        
        sc=score(n)
        if sc<record:
            continue
        if sc>record:
            record = sc
        
        geodes= r[T['geode']]
        if geodes>maxgeodes:
            maxgeodes=geodes
        
        #try building each type or nothing
        for tp in bp.keys():
            r2=copy.deepcopy(r)
            i2=copy.deepcopy(i)
            v2=copy.deepcopy(v)
            t2=t+1
            
            can=1
            for ing in bp[tp].keys():
                #print(tp,ing,bp[tp][ing])
                if i2[T[ing]] < bp[tp][ing]:
                    can=0
                    break
                    
            #harvest current robots work for next situation
            for j in range(len(i2)):
                i2[j]+=r2[j]
                v2[j]+=r2[j]
                
            if can and t2 <= maxt:
                #print("yes")
                if tp != 'nothing':
                    r2[T[tp]]+=1 
                    for ing in bp[tp].keys():
                        i2[T[ing]] -= bp[tp][ing]
                e.append((r2,i2,v2,t2))
                
        e=sorted(e,key=lambda x: score(x),reverse=True)
    #print(maxgeodes)      
    return maxgeodes

def p1():
    r=0
    bps=ld()
    for ix in bps.keys():
        mg=maxgeodes(bps[ix], 24)
        r+=ix*mg
        print(ix,mg)
        
    return r
    
print("part1:",p1(),0)

#63426 too high
#502 too low
```
