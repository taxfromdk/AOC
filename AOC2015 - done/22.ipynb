```python
import copy
import time
import json
sit = {}
sit["turn"] = "player"
sit["player_hp"] = 50
sit["player_mana"] = 500
sit["player_mana_spent"] = 0
sit["boss_hp"] = 55
sit["shield"] = 0
sit["poison"] = 0
sit["recharge"] = 0

situations = [sit]

maxmana = None

past = {}

win_mana_spent = []

def ser(s):
    ss = {}
    for k in ["turn", "player_hp", "player_mana", "boss_hp", "shield", "poison", "recharge"]:
        ss[k] = s[k]
    return json.dumps(ss)

interval = 1
nexttime = time.time() + interval
while len(situations):
    n = time.time()
    if n > nexttime:
        nexttime = n + interval
        print("%d"%(len(situations)), end='\r')
    
    s = situations[0]
    situations = situations[1:]
    
    past[ser(s)] = 1
    
    if s["shield"] > 0:
        #print("shield")
        s["shield"] -= 1

    if s["poison"] > 0:
        #print("poison")
        s["poison"] -= 1
        s["boss_hp"] -= 3

    if s["recharge"] > 0:
        #print("recharge")
        s["recharge"] -= 1
        s["player_mana"] += 101
    
    if s["player_hp"] > 0 and s["boss_hp"] > 0:                    
        if s["turn"] == "player":
            #switch turn
            s["turn"] = "boss"
            
            #Magic Missile costs 53 mana. It instantly does 4 damage.
            if s["player_mana"] >= 53:
                #print("player summons missile")
                ns = copy.deepcopy(s)
                ns["player_mana"] -= 53
                ns["player_mana_spent"]+=53
                ns["boss_hp"] -= 4                
                if not ser(ns) in past.keys():
                    if len(win_mana_spent) == 0 or min(win_mana_spent) > ns["player_mana_spent"]:
                        situations.append(ns)
                else:
                    print("dejavu")

                
            #Drain costs 73 mana. It instantly does 2 damage and heals you for 2 hit points.
            if s["player_mana"] >= 73:
                #print("player summons drain")
                ns = copy.deepcopy(s)
                ns["player_mana"] -= 73
                ns["player_mana_spent"]+=73
                ns["boss_hp"] -= 2
                ns["player_hp"] += 2
                if not ser(ns) in past.keys():
                    if len(win_mana_spent) == 0 or min(win_mana_spent) > ns["player_mana_spent"]:
                        situations.append(ns)
                else:
                    print("dejavu")

            #Shield costs 113 mana. It starts an effect that lasts for 6 turns. 
            #While it is active, your armor is increased by 7.
            if s["player_mana"] >= 113 and s["shield"] <= 0:
                #print("player buys shield")
                ns = copy.deepcopy(s)
                ns["player_mana"] -= 113
                ns["player_mana_spent"]+=113
                ns["shield"] = 6
                if not ser(ns) in past.keys():
                    if len(win_mana_spent) == 0 or min(win_mana_spent) > ns["player_mana_spent"]:
                        situations.append(ns)                    
                else:
                    print("dejavu")

            #Poison costs 173 mana. It starts an effect that lasts for 6 turns. 
            #At the start of each turn while it is active, it deals the boss 3 damage.
            if s["player_mana"] >= 173 and s["poison"] <= 0:
                #print("player summons poison")
                ns = copy.deepcopy(s)
                ns["player_mana"] -= 173
                ns["player_mana_spent"]+=173
                ns["poison"] = 6
                if not ser(ns) in past.keys():
                    if len(win_mana_spent) == 0 or min(win_mana_spent) > ns["player_mana_spent"]:
                        situations.append(ns)                    
                else:
                    print("dejavu")

                
                
            #Recharge costs 229 mana. It starts an effect that lasts for 5 turns. 
            #At the start of each turn while it is active, it gives you 101 new mana
            if s["player_mana"] >= 229 and s["recharge"] <= 0:
                #print("player buys recharge")
                ns = copy.deepcopy(s)
                ns["player_mana"] -= 229
                ns["player_mana_spent"]+=229
                ns["recharge"] = 5
                if not ser(ns) in past.keys():
                    if len(win_mana_spent) == 0 or min(win_mana_spent) > ns["player_mana_spent"]:
                        situations.append(ns)                    
                else:
                    print("dejavu")

            #hesitation has no future          
        else:
            #switch turn
            s["turn"] = "player"           
            #boss attacks
            ns = copy.deepcopy(s)
            if ns["shield"] > 0:
                ns["player_hp"] -= 8 - 7
            else:
                ns["player_hp"] -= 8
            if not ser(ns) in past.keys():
                if len(win_mana_spent) == 0 or min(win_mana_spent) > ns["player_mana_spent"]:
                    situations.append(ns)                
            else:
                print("dejavu")
                
                          
    else:
        if s["player_hp"] > 0:
            print("player won", s["player_mana_spent"])
            win_mana_spent.append(s["player_mana_spent"])
            situations = [x for x in situations if x["player_mana_spent"] < min(win_mana_spent)]
            
            pass
        elif s["boss_hp"] > 0:
            #print("boss won")
            pass
        else:
            print("both died")
            pass

print("\n\nminimum win mana:", min(win_mana_spent))


#eventually terminated at 953
```
