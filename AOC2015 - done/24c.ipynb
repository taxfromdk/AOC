```python
import copy
import time

def a(compartments):
    w=[int(x) for x in open("24.txt").readlines()]
    sw = sum(w)
    l=len(w)    
    t=int(sum(w)/compartments)

    
    minimal_space = None
    minimal_quantum = None
    minimal_situation = None

    
    
    print(w)
    print("length:%d target:%d"%(l,t))

    #quantum entanglement of all compartments but first
    def qe(s):
        r=[1 for i in range(compartments)]
        for i in range(len(s)):        
            r[s[i]] *= w[i]
        return r

    #count space used in each compartment
    def sp(s):
        r=[0 for i in range(compartments)]
        for i in s:
            r[i]+=1
        return r
    
    #compute weight of the compartments         
    def wg(s):
        ww=[0 for i in range(compartments)]
        for i in range(len(s)):
            ww[s[i]]+=w[i]
        return ww

    #cost of path from start
    def g(s):
        if not minimal_quantum is None:
            return sum(qe(s)[:-1])
        return sum(sp(s)[:-1])*50
        
    #heurestic of distance to goal
    def h(s):
        
        if not minimal_quantum is None:
            
            return sum([(minimal_quantum-x) for x in qe(s)[:-1]])
        else:
            return sum([t-x for x in wg(s)[:-1]])
        
            
    
    def f(s):
        return g(s)+h(s)
    
    def pack(s):
        r = [[] for i in range(compartments)]
        for i in range(len(s)):
            r[s[i]].append(w[i])
        return r

    def stair(a):
        for i in range(len(a) - 1):
            if(a[i] > a[i+1]):
                return False
        return True
    
    def check(s):
        #search space is huge, we must cut solution space
        weights = wg(s)
        quantum = qe(s)
        space = sp(s)
        
        

    
    
    #use hashtable for explore space, key is A* heurestic
    e={}
    
    #initial situation is all packages in last compartment 
    s=[compartments-1 for i in range(len(w))]    
    e[f(s)]=[s]
    
    counter=0

    
    nexttime=0
    
    while len(e):
        counter+=1
        
        #choose evaluation situation
        minimum_heurestic = min(e.keys())
        s=e[minimum_heurestic][0]
        e[minimum_heurestic]=e[minimum_heurestic][1:]
        if len(e[minimum_heurestic])==0:
            del e[minimum_heurestic]
        
        #print stats periodically
        n=time.time()
        if n > nexttime:
            nexttime=n+1
            print("c:%d e:%d best_qe:%d best_sp:%d"%(counter, sum([len(e[x]) for x in e.keys()]), minimal_quantum if not minimal_quantum is None else -1, minimal_space if not minimal_space is None else -1), end='\r')
        
        
        if 1:
            #walk through situation indices
            for i in range(l):
                #only move from last compartment
                if s[i] > 0:
             

                    #walk through possible target compartments
                    if 1: #for c in range(compartments-1):
                        #copy situation as it need to be put on evaluation list (hashtable)
                        ss = copy.deepcopy(s)
                        #and move packet
                        ss[i] = s[i]-1            

                        #compute weights
                        weights = wg(ss)
                        quantum = qe(ss)
                        space = sp(ss)

                        #all weights are equal - its a winner
                        if max(weights) == min(weights):
                            min_quantum = quantum[0]
                            min_space   = space[0]
                            if minimal_space is None or space[0] < minimal_space or (space[0] == minimal_space and quantum[0] < minimal_quantum):
                                minimal_quantum = quantum[0]
                                minimal_space = space[0]
                                minimal_situation = ss
                                print(counter,"new minimal qe found:", minimal_quantum, len(str(minimal_quantum)), minimal_space) 
                                print()
                        else:
                            #require compartments have weights in order
                            #and first compartment has lowest quantum
                            if stair(weights):
                                #if a solution is found we can carve solution space further
                                ok = False
                                if minimal_space is None:
                                    ok = True
                                else:
                                    if stair(space):
                                        ok = True
                                if ok:
    
                                    #compute heurestic
                                    gh=f(ss)
                                    if not gh in e.keys():
                                        e[gh] = []                                                
                                    e[gh].append(ss)

    print(compartments, " : ", minimal_quantum, pack(minimal_situation))    

print("Part 1")
     #851189229609
a(3) #10723906903
print("Part 2")
a(4)
```
