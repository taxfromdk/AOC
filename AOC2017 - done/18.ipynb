```python
def part1():
    p=[x.strip().split(" ") for x in open("18.txt").readlines()]
    
    #print(p)
    r={"snd":[]} 
    i=0
    
    def vo(x):
        if x in r:
            return r[x]
        return int(x)
    
    while i in range(len(p)):
        o=p[i]
        #print(o)
        
        if not o[1] in r:
            r[o[1]]=0
        I=o[0]
        X=o[1]
        Y=None
        if len(o)>2:
            Y=o[2]
        
        
        if I=="snd":
            r["snd"].append(r[X])
            i+=1
        elif I=="set":
            r[X]=vo(Y)
            i+=1
        elif I=="add":
            r[X]+=vo(Y)
            i+=1
        elif I=="mul":
            r[X]*=vo(Y)
            i+=1    
        elif I=="mod":
            r[X]%=vo(Y)
            i+=1
        elif I=="rcv":
            if r[X]:
                return r["snd"][-1]
            i+=1
        elif I=="jgz":
            if r[X]>0:
                i+= vo(Y)
            else:
                i+=1
        else:
            print(o)
            i+=1
        
        
"""      
snd X plays a sound with a frequency equal to the value of X.
set X Y sets register X to the value of Y.
add X Y increases register X by the value of Y.
mul X Y sets register X to the result of multiplying the value contained in register X by the value of Y.
mod X Y sets register X to the remainder of dividing the value contained in register X by the value of Y (that is, it sets X to the result of X modulo Y).
rcv X recovers the frequency of the last sound played, but only when the value of X is not zero. (If it is zero, the command does nothing.)
jgz X Y jumps with an offset of the value of Y, but only if the value of X is greater than zero. (An offset of 2 skips the next instruction, an offset of -1 jumps to the previous instruction, and so on.)
"""
    
    

print("part1:", part1(), 2951)




```

```python
inbox=[[],[]]
counters=[0,0]

def prog(id):
    
    p=[x.strip().split(" ") for x in open("18.txt").readlines()]
    
    #print(p)
    r={'p':id} 
    i=0
    #print("I am ",r["p"])
    
    
    def vo(x,r):
        
        try:
            return int(x)
        except ValueError:
            return r[x]
        
    
    while i >= 0 and i < len(p):
        o=p[i]
        #print(id, o)
        
        
            
        I=o[0]
        X=o[1]
        
        if not X in r and not X.isnumeric():
            #print(id, "reg", o[1])
            r[o[1]]=0
        
        
        Y=None
        if len(o)>2:
            Y=vo(o[2],r)
        
        
        if I=="snd":
            inbox[1-id].append(r[X])
            counters[id]+=1
            #print(id, "sent", r[X])
            
            i+=1
        elif I=="set":
            r[X]=vo(Y,r)
            i+=1
        elif I=="add":
            r[X]+=vo(Y,r)
            i+=1
        elif I=="mul":
            r[X]*=vo(Y,r)
            i+=1    
        elif I=="mod":
            r[X]%=vo(Y,r)
            i+=1
        elif I=="rcv":
            while len(inbox[id])==0:
                yield 0
            r[X]=inbox[id][0]
            inbox[id]=inbox[id][1:]
            #print(id,"recieving",r[X])
            i+=1
        elif I=="jgz":
            if vo(X,r) > 0:
                i+= vo(Y,r)
                #print(Y)
            else:
                i+=1
        else:
            print(o)
            i+=1
        yield 1
        
        #for k in r.keys():
        #    assert(type(r[k]) is int)
        
    while 1:
        yield 0
    
   
      
    
    

def part2():
    pa=prog(0)
    pb=prog(1)
    c=1
    while next(pb) or next(pa):
        if c%100000==0:
            print(c,counters,end="\r")
        c+=1
    print()
    return counters[1]
    


print("part2:", part2(), 7366)

```
