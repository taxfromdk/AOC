```python
import re

def prep(l):
    r = re.findall("[a-z]+|[+-<>=!]+|[-0-9]+",l)
    return r

def nm(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

def gv(i, r):
    if nm(i):
        return int(i)
    return r[i]
    
def e(a,c,b,r):
    a = gv(a,r)
    b = gv(b,r)
    
    if c == '==':
        return a == b
    
    elif c == '<':
        return a < b
    elif c == '<=':
        return a <= b
    
    elif c == '>':
        return a > b
    elif c == '>=':
        return a >= b
    
    elif c == '!=':
        return a != b
    
    else:
        print("evaluation", c)
    print("fail eval")
    return True
        

def part(part):
    p=[prep(x.strip()) for x in open("08.txt").readlines()]
    r={}
    i=0
    allhigh = 0
    while i<len(p):
        #print(p[i])
        reg= p[i][0]
        op = p[i][1]
        v  = p[i][2]
        #if
        a  = p[i][4]
        cnd= p[i][5]
        b  = p[i][6]
        
        for z in [reg, v, a, b]:
            if not nm(z) and z not in r:
                #print("adding reg:", z)
                r[z]=0
                
        if e(a, cnd, b, r):
            if op == "dec":
                r[reg] -= gv(v,r)
            elif op == "inc":
                r[reg] += gv(v,r)
            else:
                print("unknown",op)
        
        allhigh = max(allhigh, max(r.values()))
        i+=1
    
    if part == 2:
        return allhigh
    return max(r.values())



#3325
print("part1:", part(1),4416)
print("part2:", part(2),5199)

```
