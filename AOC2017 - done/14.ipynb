```python
def kh(i):
    seq=[ord(x) for x in i]+[17, 31, 73, 47, 23]
    #print(i)
    
    L=list(range(256))
    n=0
    skip_steps=0
    for _ in range(64):
        
        for l in seq:
            if l>256:
                continue
            
        
            t=[]
            b=n
            for j in range(l):
                t.append(L[b])
                b=(b+1)%256
            
            #insert reverse and move
            for v in t[::-1]:
                L[n]=v
                n=(n+1)%256
            n=(n+skip_steps)%256
            skip_steps+=1
    o=""
    while len(L):
        l=L[:16]
        L=L[16:]
        a=None
        for x in l:
            if a is None:
                a=x
            else:
                a ^= x
        o+="%02x"%(a)
    return o




def regions(B):
    cnt=0
    while 1:
        done=1
        for ix in range(128*128):
            x=ix%128
            y=ix//128
            if B[y][x]:
                done=False
                f=[[x,y]]
                
                cnt+=1
                while len(f):
                    n=f[0]
                    f=f[1:]
                    
                    xx,yy=n
                    if B[yy][xx]:
                        B[yy][xx]=0
                        for dx,dy in [[-1,0], [1,0], [0,-1], [0,1]]:
                            xxx=xx+dx
                            yyy=yy+dy
                            if xxx >= 0 and xxx < 128 and yyy >= 0 and yyy < 128:
                                #print(xxx,yyy)
                                if B[yyy][xxx]:
                                
                                    f.append([xxx,yyy])
        if done:
            return cnt

def calc(i):
    B=[]
    #for l in b:
    #    print(l)
    for j in range(128):
        s="%s-%d"%(i,j)
        h=kh(s)
        #b=bin(int(h,16))[2:]
        b=format(int(h,16), '0128b')
        
        #print(len(b),b)
        #print(s,h,b)
        B.append([int(x) for x in list(b)])
    #print(B)
    #print([len(l) for l in B])
    used =sum([sum(y) for y in B])

    
    reg=regions(B)
    
    return used,reg

print("test:", calc("flqrgnkx"),8108,1242)
      
print("part:", calc("jzgqcdpd"),8074,'?')
      
```
