```python
import gc
import time

D={
    'U':( 0,-1),
    'R':( 1, 0),
    'D':( 0, 1),
    'L':(-1, 0)
}


dirs={
    'U': "LUR",
    'R': "URD",
    'D': "RDL",
    'L': "DLU"
}

def axis(d):
    if d in "LR":
        return 0
    return 1

def state(s):
    x,y,d,hl,stp,ts=s
    return (x,y,axis(d),stp)
    
def p1(fn):
    
    M=[[int(x) for x in l] for l in open(fn).read().split('\n')]
    
    w=len(M[0])
    h=len(M)
    M2={}
    for y in range(h):
        for x in range(w):
            M2[(x,y)]=M[y][x]
    M=M2
    
    
    
    
    #print(w,h)
    
    goal=(w-1,h-1)
    #pos, direction, accumulated heatloss, steps, total steps
    S={}
    S[0]=[(0, 0, 'R', 0, 0, 0),(0,0,'D',0,0,0)]
    O={}
    i=0
    lt=time.time()
    li=0
    while 1:
        si=min(S.keys())
        #keep S as dict of lists to avoid sort
        s=S[si][0]
        S[si]=S[si][1:]
        if len(S[si])==0:
            del S[si]
            
        if state(s) in O:
            continue
        
        x,y,d,hl,stp,ts=s
        ts2=ts+1
        i+=1
        t=time.time()
        if t > lt + 1:
            print("\r",i,ts,sum([len(S[k]) for k in S.keys()]), i-li,"       ",end="")
            lt=t
            li=i
            #gc.collect()
             
            
        if (x,y)==goal:
            return hl
        
        #mark visited
        O[state(s)]=1
        
        #try forward,left,right
        for d2 in dirs[d]:
            dx,dy=D[d2]
            x2=x+dx
            y2=y+dy
            if x2>= 0 and x2<w:
                if y2>=0 and y2<h:
                    
                    hl2=hl+M[(x2,y2)]
                    stp2=stp+1
                    if d!=d2:
                        stp2=1
                    if stp2<4:
                        s2=(x2,y2,d2,hl2,stp2,ts2)
                        if not state(s2) in O:
                            if not hl2 in S:
                                S[hl2]=[]
                            S[hl2].append(s2)
                        
                     
    return None


print("p1:",p1("17.txt"),1260)


def state2(s):
    #return state(s)
    x,y,d,hl,stp,ts=s
    return (x,y,axis(d))
      

def p2(fn):
    
    M=[[int(x) for x in l] for l in open(fn).read().split('\n')]
    
    w=len(M[0])
    h=len(M)
    M2={}
    for y in range(h):
        for x in range(w):
            M2[(x,y)]=M[y][x]
    M=M2
    
    
    
    
    #print(w,h)
    
    goal=(w-1,h-1)
    #pos, direction, accumulated heatloss, steps, total steps
    S={}
    S[0]=[(0, 0, 'R', 0, 0, 0),(0,0,'D',0,0,0)]
    O={}
    i=0
    lt=time.time()
    li=0
    while 1:
        si=min(S.keys())
        #keep S as dict of lists to avoid sort
        s=S[si][0]
        S[si]=S[si][1:]
        if len(S[si])==0:
            del S[si]
            
        if state(s) in O:
            continue
        
        x,y,d,hl,stp,ts=s
        ts2=ts+1
        i+=1
        t=time.time()
        if t > lt + 1:
            print("\r",i,ts,sum([len(S[k]) for k in S.keys()]), i-li,"       ",end="")
            lt=t
            li=i
            #gc.collect()
             
            
        if (x,y)==goal and stp >= 3 and stp<10:
            return hl
        
        #mark visited
        O[state2(s)]=1
        
        #try forward,left,right
        for d2 in dirs[d]:
            dx,dy=D[d2]
            x2=x+dx
            y2=y+dy
            if x2>= 0 and x2<w:
                if y2>=0 and y2<h:
                    allow=None
                    if d==d2:
                        allow=stp<9
                    else:
                        allow=stp>=3
                        
                    hl2=hl+M[(x2,y2)]
                    stp2=stp+1
                    
                    
                    if d!=d2:
                        stp2=0
                    
                    if allow:
                        s2=(x2,y2,d2,hl2,stp2,ts2)
                        if not state2(s2) in O:
                            if not hl2 in S:
                                S[hl2]=[]
                            S[hl2].append(s2)
                        
                     
    return None


#1423 too high
#1340 too low
#1399
#1405 not
#1406 not
#1407 not
print("p2:",p2("17.txt"),2)
```
