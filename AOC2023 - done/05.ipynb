```python
def parse(fn):
    D=open(fn).read().split("\n\n")
    
    M={}
    for d in D:
        #print(d)
        d=d.split(":")
        M[d[0].split(" ")[0]]=[[int(y) for y in x.split(" ") if len(y)] for x in d[1].split("\n") if len(x)]
    
    return M

def cnv(M, n, i):
    for x in M[n]:
        #print(x,i)
        dst=x[0]
        src=x[1]
        l=x[2]
        if i >= src and i < src+l:
            return i-src+dst
    return i

def part1(fn):
    M=parse(fn)
    lloc=None
    for seed in M["seeds"][0]:
        #print(seed)
        soil=cnv(M,"seed-to-soil", seed)
        fert=cnv(M,"soil-to-fertilizer",soil)
        wate=cnv(M,"fertilizer-to-water",fert)
        ligh=cnv(M,"water-to-light",wate)
        temp=cnv(M,"light-to-temperature",ligh)
        humi=cnv(M,"temperature-to-humidity",temp)
        loca=cnv(M,"humidity-to-location",humi)
        if lloc is None or loca < lloc:
            lloc=loca
    return lloc

def smp(r):
    l=len(r)
    for i in range(l):
        for j in range(l):
            if i!=j:
                A,B=r[i]
                a,b=r[j]
                assert(a<b)
                assert(A<B)
                if A>b+1 or B<a-1:#no touch
                    pass
                else:
                    #print("simplify",a,b,A,B)
                    r2=[]
                    r2.append((min(a,A),max(b,B)))
                    for k in range(l):
                        if not k in [i,j]:
                            r2.append(r[k])
                        
                    return sorted(r2, key=lambda x: x[0]),0
    return r,1

def simplify(r):
    #return list(set(r))
    l=len(r)
    while 1:
        r,d=smp(r)
        if d:
            break
    #print("simplifying",l-len(r))
    return r
    


def part2(fn):
    M=parse(fn)
    #print(M)
    r=[]
    S=M["seeds"][0]
    #print(S)
    for i in range(0,len(S),2):
        #print(i)
        s=S[i+0]
        l=S[i+1]
        r.append((s,s+l-1))
    for k in M.keys():
        if k != "seeds":
            print(k,len(r))
            r2=[]
            while len(r):
                a,b=r.pop()
                assert(a<b)
                h=0
                for x in M[k]:
                    dst,A,l=x
                    B=A+l-1
                    o=dst-A
                    assert(A<B)
                    
                    if b < A or B < a: #miss
                        pass
                        #r2.append((a,b))
                        #print("miss",r2[-1])
                    else:
                        h=1
                        r2.append((max(A,a)+o,min(B,b)+o))
                        #if r2[-1][0]==0:
                        #    print(A,B)
                        #print("touch",r2[-1])
                        assert(r2[-1][0] < r2[-1][1])
                        if a<A:
                            r.append((a,A-1))
                            #print("l",r2[-1])
                        if b>B:
                            r.append((B+1,b))
                            #print("r",r2[-1])
                if h==0:
                    r2.append((a,b))
                    
            r=simplify(r2)
            for z in r:
                pass
                #print(z)
    lloc= None
    for a,b in r:
        if lloc== None or lloc>a:
            lloc=a
    return lloc
        
                      
    

print("part1:",part1("05.txt"),214922730)
print("part2:",part2("05.txt"),148041808)


```
